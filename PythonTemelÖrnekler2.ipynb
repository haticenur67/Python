```python
#Bir liste oluşturun

liste = ["mavi","kirmizi","yesil","sari"]

print(liste)
```

```python
#Erişim listesi öğeleri

liste = ["mavi","kirmizi","yesil","sari"]

print(liste[0])
print(liste[-1])
print(liste[1:4])
```

```python
#Bir liste öğesinin değerini değiştirin

liste = ["mavi","sari","yesil","kirmizi","siyah"]
print(liste)

print("yeni hali")
liste[4] = "turuncu"
print(liste)
```

```python
#Bir listede döngü oluşturun

liste = ["kirmizi","mavi","sari","yesil","mor"]
for i in liste:
  print(i)
```

```python
#Bir liste öğesinin mevcut olup olmadığını kontrol edin

liste = ["kirmizi","mavi","sari","yesil","mor"]
for i in liste:
  if i in liste:
    print(f"{i} listede mevcut")
```

```python
#Bir listenin uzunluğunu alın

liste = ["kirmizi","mavi","sari","yesil","mor"]
print(len(liste))
```

```python
#Bir öğeyi listenin sonuna ekleyin

liste = ["kirmizi","mavi","sari","yesil","mor"]
liste.insert(len(liste),"turuncu")
print(liste)
```

```python
#Belirtilen dizine bir öğe ekleyin

liste = ["kirmizi","mavi","sari","yesil","mor"]
liste.append("pempe")
print(liste)
```

```python
#Bir öğeyi kaldırın

liste = ["kirmizi","mavi","sari","yesil","mor"]
liste.remove("kirmizi")
print(liste)
```

```python
#Son öğeyi kaldırın

liste = ["kirmizi","mavi","sari","yesil","mor"]
a = len(liste) - 1
del liste[a]
print(liste)
```

```python
#Belirtilen bir dizindeki bir öğeyi kaldırın

liste = ["kirmizi","mavi","sari","yesil","mor"]
liste.pop(4)
print(liste)
```

```python
#Bir listeyi boşaltın

liste = ["kirmizi","mavi","sari","yesil","mor"]
liste.clear()
print(liste)
```

```python
#Bir liste oluşturmak için list() kurucusunu kullanma

liste = ["kirmizi","mavi","sari","yesil","mor"]
yeniliste = list(liste)
print(yeniliste)
```

```python
#Bir tuple oluşturun

newtuple = ("kirmizi","mavi","sari")
print(newtuple)
```

```python
#Erişim tuple öğeleri

newtuple = ("kirmizi","mavi","sari","mor","yesil")
print(newtuple[3])
print(newtuple[-1])
print(newtuple[:3])
```

```python
#Tuple değerlerini değiştir

newtuple = ("kirmizi","mavi","sari","mor","yesil")
a = list(newtuple)
a[2] = "pembe"
newtuple = tuple(a)
print(newtuple)
```

```python
#Bir tuple'da döngü

newtuple = ("kirmizi","mavi","sari","mor","yesil")
for i in newtuple:
  print(i)
```

```python
#Bir tuple öğesinin var olup olmadığını kontrol edin

newtuple = ("kirmizi","mavi","sari","mor","yesil")
for i in newtuple:
  if i in newtuple:
    print(f"{i} bu tupleda var")
```

```python
#Bir tuple'ın uzunluğunu alın

newtuple = ("kirmizi","mavi","sari","mor","yesil")
print(len(newtuple))
```

```python
#Bir tuple'ı sil

newtuple = ("kirmizi","mavi","sari","mor","yesil")
del newtuple
```

```python
#Bir tuple oluşturmak için tuple() oluşturucusunu kullanmak

liste = ["kirmizi","mavi","sari","yesil","mor"]
newtuple = tuple(liste)
print(newtuple)
```

```python
#Bir sözluk oluşturun

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}
print(sozluk)
```

```python
#Bir sözlüğün öğelerine erişin

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}
print(sozluk["isim"])

a = sozluk.get("yas")
print(a)
```

```python
#Sözlükteki belirli bir öğenin değerini değiştirin

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}
sozluk["sehir"] = "ankara"
print(sozluk)
```

```python
#Sözlülteki tüm anahtar adlarını tek tek yazdırın

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}
a = sozluk.keys()
print(a)
```

```python
#Sözlükteki tüm değerleri tek tek yazdırın

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}
print(sozluk)
```

```python
#Bir sözlüğün değerlerini döndürmek için values() fonksiyonunu kullanma

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}
a = sozluk.values()
print(a)
```

```python
#items() fonksiyonunu kullanarak hem anahtarlar hem de değerler arasında döngü kurun

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}
a = sozluk.items()
print(a)
```

```python
#Bir anahtarın var olup oladığını kontrol edin

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}

if "yas" in sozluk:
  print("yas anahtari bu sozlukte bulunuyor")
else:
  print("yas anahtari bu sozlukte bulunmuyor")
```

```python
#Bir sözlüğün uzunluğunu alın

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}

print(len(sozluk))
```

```python
#Bir öğeyi sözlüğe ekleyin

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}

sozluk["meslek"] = "ogretmen"
print(sozluk)
```

```python
#Bir öğeyi sözlükten kaldırın

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}

sozluk.pop("yas")
print(sozluk)
```

```python
#Bir sözlüğü boşaltın

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}

del sozluk
```

```python
#Bir sözlük oluşturmak için dict() oluşturucusunu kullanmak

sozluk = {
    "isim" : "ayse",
    "yas" : 25,
    "sehir" : "istanbul"
}

yenisozluk = dict(sozluk)
print(yenisozluk)
```
