```python
#8. Armstrong Sayısını Kontrol Etmek İçin Python Programı
sayi = int(input("sayi girin:"))
a = sayi
i=0
while a > 0:
  i = i+1
  a //= 10

a = sayi
top = 0

while a > 0:
  b = a % 10
  top += b ** i
  a //= 10

if top == sayi:
  print("amstrong sayidir")
else:
  print("amstrong sayi degildir")

```

```python
#9. Bir Aralıkta Armstrong Sayısını Bulan Python Programı
min = int(input("baslangic degeri:"))
max = int(input("bitis degeri:"))

for i in range(min,max+1):
  a=i
  sayac = 0
  while a > 0:
    sayac = sayac+1
    a //= 10

  top = 0
  a=i
  while a > 0:
    b = a % 10
    top += b ** sayac
    a //= 10

  if top == i:
    print(f"{i}")
```

```python
#12. Özyineleme Kullanarak Fibonacci Dizisini Gösteren Python Programı

def fonk(sayi):
  if sayi <= 0:
    return 0
  elif sayi == 1:
    return 1
  else:
    return fonk(sayi-1)+fonk(sayi-2)

n = int(input("terim sayisi:"))

print(f"{n} terimlik fibonacci dizisi:")
for i in range(n):
  print(fonk(i))
```

```python
#13. Özyineleme Kullanarak Sayının Faktöriyelini Bulan Python Programı
def fonk(n):
  if n <= 0:
    return 0
  elif n == 0 or n == 1:
    return 1
  else:
    return n*fonk(n-1)

sayi = int(input("sayiyi girin:"))
print(f"{sayi}! : {fonk(sayi)}")
```

```python
#14. İki Matrisi Toplamak İçin Python Programı
satir = int(input("satir sayisi:"))
sutun = int(input("sutun sayisi:"))

print("ilk matrisi girin")
matris1 = []
for i in range(satir):
  a = list(map(int,input(f"{i+1}.satiri girin: ").split()))
  matris1.append(a)

print("ikinci matrisi girin")
matris2 =[]
for i in range(satir):
  a = list(map(int,input(f"{i+1}.satiri girin: ").split()))
  matris2.append(a)

top = []
for i in range(satir):
  a = []
  for k in range(sutun):
    a.append(matris1[i][k] + matris2[i][k])
  top.append(a)

print("matrislerin toplami")
for a in top:
  print(a)
```

```python
#15. Bir Matrisi Transpoze Etmek İçin Python Programı

satir = int(input("satir sayisi: "))
sutun = int(input("sutun sayisi: "))

matris = []
for i in range(satir):
  a = list(map(int,input(f"{i+1}.satiri girin: ").split()))
  matris.append(a)

transpoze = []
for i in range(sutun):
  b = []
  for k in range(satir):
    b.append(matris[k][i])
  transpoze.append(b)

print("yeni matris")
for j in transpoze:
  print(j)
```

```python
#16. İki Matrisi Çarpmak İçin Python Programı
matris1 = [
    [1,2,3],
    [4,5,6]
]
matris2 = [
    [2,3],
    [4,5],
    [6,7]
]

if len(matris1[0]) != len(matris2):
  print("boyut hatasi")
else:
  carpim = []
  for i in range(len(matris1)):
    a = []
    for k in range(len(matris2[0])):
      top = 0
      for j in range(len(matris2)):
        top += matris1[i][j] * matris2[j][k]
      a.append(top)
    carpim.append(a)

print("sonuc: ")
for b in carpim:
  print(b)
```
